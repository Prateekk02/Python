## <b><i>Graphical Analysis

### Using Graphical Analysis to analyse the correlation.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,8))    # controls the size of the figure.
sns.heatmap(df.corr(), annot=True)   # In the heatmap denotes the correlation of the feature wrt to other along with values when annot is set to True.
```

```python
df.quality.value_counts().plot(kind= 'bar')
```

### To observe the distribution of any feature.

```python
sns.distplot(df['fixed acidity'])
```

```python
sns.histplot(df['fixed acidity'], kde=True)
```

### To get all the distribution in one graph

```python
for i in df.columns:
    sns.histplot(df[i], kde=True)
```

### Categorical Plot

- Also helps us to find outlier

```python
sns.catplot(x='quality' , y='alcohol', data=df, kind='box')
```

```python
sns.scatterplot(x='alcohol', y='pH', hue='quality',data=df)
```
