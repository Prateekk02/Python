```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("This is x data")
plt.ylabel("This is y data")
plt.title("This is title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
colours = np.random.rand(50)
sizes = 100*np.random.rand(50)
plt.scatter(x,y, c = colours,s = sizes)
```

```python
plt.scatter(x,y, c = colours,s = sizes, alpha = .5)   # alpha denotes opacity.
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
plt.bar(x,y)
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
plt.barh(x,y)
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
plt.figure(figsize=(6,4))   # Control the size of figure.
plt.grid()   # For grid in background
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph")
plt.plot(x,y, '--r')   # --b for blue dashed lines.

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)


plt.scatter(x,y , c = 'purple' , s = 500, alpha = .8, marker = 'x')
```

```python
x = [3,4,5,6,7]
y = [4,5,6,7,8]

plt.plot(x,y)
```

```python
data = [1,2,3,4,5,1,3,6,7,8,1,2,3]
plt.hist(data)     # Tells the frequency of the elements in the data.
```

```python
x = [4,5,6,7,8,9]
y = [60,30,80,19,40,1]
colour = ['red', 'blue', 'green', 'purple', 'orange','yellow']
plt.scatter(x,y, c= colour)
plt.show()
```

#### Generating 3 D plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection= '3d')
ax.scatter(x,y,z)
plt.show()
```
