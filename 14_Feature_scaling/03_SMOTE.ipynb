# SMOTE(Sythetic Minority Oversampling Technique)

### Making an imbalance dataset using sklearn

```python
from sklearn.datasets import make_classification 
# X is independent feature
# Y is dependent feature
X,Y = make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class=1, weights=[0.90],
                    random_state=1)
```

### Making dataframe for X and Y

```python
import pandas as pd

df1 = pd.DataFrame(X, columns=['f1','f2'])
df2 = pd.DataFrame(Y, columns=['target'])

final_df = pd.concat([df1, df2], axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

### Visualising the dataset using graph

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c= final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

### Transforming the dataset

```python
oversample = SMOTE()
import numpy as np
X_new,Y_new = oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```

```python
X_new.shape,Y_new.shape
```

### Visualising 

```python
df3 = pd.DataFrame(X_new, columns=['f1','f2'])
df4 = pd.DataFrame(Y_new, columns=['target'])
```

```python
oversample_df = pd.concat([df3,df4], axis=1)
```

```python
plt.scatter(oversample_df['f1'], oversample_df['f2'],c=oversample_df['target'])
```
